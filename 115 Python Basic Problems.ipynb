```python
#18. Quadratic Equation Solver:
#Write a Python program that takes the coefficients (a, b, c) of a quadratic equation as input and calculates and prints the real roots (if they exist) or a message indicating the complex roots.

import cmath

a = float(input("Write the coefficient a: "))
b = float(input("Write the coefficient b: "))
c = float(input("Write the coefficient c: "))

def quadratic_roots(a, b, c):

    discriminant = b**2 - 4*a*c

    # Check the nature of roots
    if discriminant > 0:
        # Real and distinct roots
        root1 = (-b + (discriminant)**0.5) / (2*a)
        root2 = (-b - (discriminant)**0.5) / (2*a)
        return root1, root2
    elif discriminant == 0:
        # Real and equal roots
        root = -b / (2*a)
        return root,
    else:
        # Complex roots
        root1 = (-b + cmath.sqrt(discriminant)) / (2*a)
        root2 = (-b - cmath.sqrt(discriminant)) / (2*a)
        return root1, root2


if a != 0:
    roots = quadratic_roots(a, b, c)
    print("The roots of the quadratic equation are:", roots)
else:
    print("Invalid input. 'a' should not be zero for a quadratic equation.")


```

```python
def multiply_matrices(matrix1, matrix2):

    rows1, cols1 = len(matrix1), len(matrix1[0])
    rows2, cols2 = len(matrix2), len(matrix2[0])

    if cols1 != rows2:
        return "Matrices cannot be multiplied. Number of columns in the first matrix must be equal to the number of rows in the second matrix."

    result_matrix = [[0 for _ in range(cols2)] for _ in range(rows1)]

    for i in range(rows1):
        for j in range(cols2):
            for k in range(cols1):
                result_matrix[i][j] += matrix1[i][k] * matrix2[k][j]

    return result_matrix

# Example Usage:
# Matrices for illustration (you can replace them with your own matrices)
matrix1 = [[1, 2, 3], [4, 5, 6]]
matrix2 = [[7, 8], [9, 10], [11, 12]]

result = multiply_matrices(matrix1, matrix2)

if isinstance(result, str):
    print(result)
else:
    print("Matrix 1:")
    for row in matrix1:
        print(row)

    print("\nMatrix 2:")
    for row in matrix2:
        print(row)

    print("\nResultant Matrix:")
    for row in result:
        print(row)
```

```python
def add_matrices(matrix1, matrix2):
    """
    Add two matrices represented as nested lists.
    """
    rows1, cols1 = len(matrix1), len(matrix1[0])
    rows2, cols2 = len(matrix2), len(matrix2[0])

    if rows1 != rows2 or cols1 != cols2:
        return "Matrices must have the same dimensions for addition."

    result_matrix = [[0 for _ in range(cols1)] for _ in range(rows1)]

    for i in range(rows1):
        for j in range(cols1):
            result_matrix[i][j] = matrix1[i][j] + matrix2[i][j]

    return result_matrix

# Example Usage:
# Matrices for illustration (you can replace them with your own matrices)
matrix1 = [[1, 2, 3], [4, 5, 6]]
matrix2 = [[7, 8, 9], [10, 11, 12]]

result = add_matrices(matrix1, matrix2)

if isinstance(result, str):
    print(result)
else:
    print("Matrix 1:")
    for row in matrix1:
        print(row)

    print("\nMatrix 2:")
    for row in matrix2:
        print(row)

    print("\nResultant Matrix:")
    for row in result:
        print(row)
```

```python
def flatten_nested_list(nested_list):
    """
    Flatten a nested list and convert it into a single-dimensional list.
    """
    flat_list = []

    for element in nested_list:
        if isinstance(element, list):
            flat_list.extend(flatten_nested_list(element))
        else:
            flat_list.append(element)

    return flat_list

# Example Usage:
try:
    # Taking input nested list from the user
    nested_list_str = input("Enter a nested list (e.g., [[1,2],[3,4]]): ")

    # Using eval to convert the input string to a nested list
    nested_list = eval(nested_list_str)

    result = flatten_nested_list(nested_list)
    print("Flattened List:", result)
except (ValueError, SyntaxError):
    print("Invalid input. Please enter a valid nested list.")
```

```python
#67. Diagonal Sum of Matrix:
#Given a square matrix represented as a nested list, write a Python program to calculate the sum of the elements in the main diagonal.


matrix_str = input("Enter a square matrix represented as a nested list: ")


def sum_main_diagonal(matrix):

    if not matrix or len(matrix) != len(matrix[0]):
       # Not a square matrix
       return None

    diagonal_sum = sum(matrix[i][i] for i in range(len(matrix)))
    return diagonal_sum


#convert the input string to a square matrix
matrix = eval(matrix_str)

diagonal_sum = sum_main_diagonal(matrix)
if diagonal_sum is not None:
    print("Original Square Matrix:")
    for row in matrix:
        print(row)
    print("\nSum of Main Diagonal Elements:", diagonal_sum)
else:
    print("Invalid input. Please enter a valid square matrix.")
```

```python
#68. Nested List Element Search:
#Write a Python program to search for a specific element in a nested list and return its position (row and column indices).

def search_element(matrix, target):

    for i, row in enumerate(matrix):
        for j, element in enumerate(row):
            if element == target:
                return i, j
    return None
    # Element not found


matrix_str = input("Enter a nested list representing a matrix: ")

matrix = eval(matrix_str)

target_element = eval(input("Enter the element to search for: "))

position = search_element(matrix, target_element)
if position is not None:
    print("Original Matrix:")
    for row in matrix:
        print(row)
    print(f"\nPosition of {target_element}: Row {position[0]}, Column {position[1]}")
else:
    print(f"Element {target_element} not found in the matrix.")
```

```python
#92. Prime Number Checker:
#Write a Python program that takes a number as input and determines if it is a prime number or not. Use a `for` loop to check for factors. If a factor is found, `break` out of the loop.

no=int(input("Enter number to check Prime or Not: "))

def check_Prime(num):
    flag = False

    if num == 1:
        print(num, "is not a prime number")
    elif num > 1:
        # check for factors
        for i in range(2, num):
            if (num % i) == 0:
                # if factor is found, set flag to True
                flag = True
                # break out of loop
                break

        # check if flag is True
        if flag:
            return False
        else:
            return True

print(check_Prime(no))
```
